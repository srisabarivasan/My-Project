## Data cleaning

```python
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
column=['EngineSize', 'Cylinders', 'Horsepower', 'MPG_City', 'MPG_Highway',
       'Weight', 'Wheelbase', 'Length']
for col in column:
    df[col].fillna(df[col].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```
